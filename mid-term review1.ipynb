### Task

Construct an immunized portfolio using the given bonds and obligation cash flow stream. 

1. Calculate the present value of the obligation.
2. Calculate the quasi-modified duration of the obligation.
3. Calculate the quasi-modified duration of Bond 1 and Bond 2.
4. Solve the system of equations to find the appropriate weights for the bonds in the portfolio.

```python
#immunization and quasi-modified duration of obligation and bond1,2
import numpy as np

# Yearly spot rates
s = np.array([7.20, 8.10, 8.70, 9.25, 9.70, 10.10, 10.45, 10.80, 11.10, 11.38, 11.62, 11.83]) / 100

# Obligation cash flow stream
a = np.array([400, 800, 700, 600, 200, 150, 150, 100])

# Bond information
c1 = 5
c2 = 12
F = 100
T1 = 10
T2 = 6

#cash flows of the bonds
flow1 = np.concatenate([c1 * np.ones(T1 - 1), [F + c1]])
flow2 = np.concatenate([c2 * np.ones(T2 - 1), [F + c2]])

def present_value(cash_flow, spot_rates):
    """Calculate the present value of a cash flow stream given the spot rates."""
    n = len(cash_flow)
    pv = 0
    
    for i in range(n):
        discount_factor = (1 + spot_rates[i]) ** -((i + 1))
        pv += cash_flow[i] * discount_factor
        
    return pv

def quasi_modified_duration(cash_flow, spot_rates):
    """Calculate the quasi-modified duration of a cash flow stream (for m=1) given the spot rates."""
    n = len(cash_flow)
    pv = present_value(cash_flow, spot_rates)
    duration = 0
    
    for i in range(n):
        time_weighted_cash_flow = (i + 1) * cash_flow[i]
        discount_factor = (1 + spot_rates[i]) ** (-(i + 2))
        duration += time_weighted_cash_flow * discount_factor
        
    return duration / pv

# Calculate the present value and quasi-modified duration of the obligation
PV = present_value(a, s)
D = quasi_modified_duration(a, s)


# Calculate the present value and quasi-modified duration of Bond 1
P1=present_value(flow1, s)
D1 = quasi_modified_duration(flow1, s)

# Calculate the present value and quasi-modified duration of Bond 2
P2=present_value(flow2, s)
D2 = quasi_modified_duration(flow2, s)

# Solve the system of equations
A = np.array([[P1, P2], [P1 * D1, P2 * D2]])
B = np.array([PV, PV * D])
X = np.linalg.solve(A, B)

print("Weights for the immunized portfolio:")
print(f"Bond 1: {X[0]:.2f}")
print(f"Bond 2: {X[1]:.2f}")
```
